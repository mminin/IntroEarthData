The following example (courtesy A. Bonanni, ECMWF) creates a global map of the temperature at 2m height, dyncamically queried.

#### Load required libraries

````python
import time
import datetime
import requests
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
````

#### Helper functions

````python
def plot_global_2d_data(_json_map, 
                        plot_title=None, 
                        data_bias=None, 
                        unit_measure=None,
                        col_min=None,
                        col_max=None,
                        save_path=None):
    
    # get the data and add a bias, if necessary
    nn = np.asarray(_json_map)
    if data_bias:
        nn += data_bias
    
    # get the mesh grid
    mmx, mmy = np.meshgrid(np.arange(nn.shape[0])/2.0-180, np.arange(nn.shape[1], 0, -1)/2-90 )

    # set-up the plotimage
    plt.close()
    fig = plt.figure(figsize=(16,8))
    
    if plot_title:
        plt.title(plot_title)

    my_cmap = "jet"
    vmin=col_min if col_min else nn.min()
    vmax=col_max if col_max else nn.max()
    plt.pcolormesh(mmx.transpose(), mmy.transpose(), nn, cmap=my_cmap, vmin=vmin, vmax=vmax)
    if unit_measure:
        cb = plt.colorbar()
        cb.set_label(unit_measure, labelpad=5, y=0.45)        
        cb.cmap.set_over('k')

    # axes configuration
    plt.axis([-180,180,-90,90])
    plt.xticks(np.linspace(-180,180,13))
    plt.yticks(np.linspace(-90,90,13))
    plt.xlabel("Longitude")
    plt.ylabel("Latitude")
    
    if save_path:
        plt.savefig(save_path)

    # do show()
    plt.show()
````

````python
url = "http://earthserver.ecmwf.int/rasdaman/ows?service=WCS&version=2.0.1&"+ \
"request=ProcessCoverages&query=for%20c%20in%20(temp2m)%20return%20encode"+ \
"(c[Lat(-90:90),Long(-180:179.5),ansi(%222003-02-01T12:00%22)],%22json%22)"

_response = requests.get(url, stream=True)
_response.raise_for_status()
json_data = _response.json()

plot_global_2d_data(json_data, 
                    plot_title="Temperature at h=2m - 01 Aug 2003", 
                    data_bias=-273.15,
                    unit_measure="$^\circ$C",
                    save_path="./output/global_t2m.png")
````

````python
#Subset using SoilMoisture dataset
#gat a daily map
#t in UNIX time
from IPython.display import Image

filename = "test_0.png"

coverage_file = my_wcs.getCoverage(
    identifier=['SoilMoisture_4326_01'],
    format='image/png',
    subsets=[('t',"1508716800","1508803199")]
)

with open( filename, 'wb') as outfile:
    outfile.write(coverage_file.read())

Image(filename = filename)
````

### Retrieve the data URL and check

````python
data_url = 'https://doi.pangaea.de/10.1594/PANGAEA.855693?format=text' # replace this with your url
#get data from the url using package requests
import requests
r=requests.get(data_url)
data=r.text
print(data[:4000]) #see what's inside
````

````python
#saving the data
with open('../data/boknis.txt', 'w+') as outfile: outfile.write(data)
````

````python
# Header on Pangaea is marked with /* HEADER */
# split data and header:
h, d = data.split('*/')
````

````python
print(h)
````

````python
print(d[:400])
````

````python
#parse data into a list
lines=d.split('\n')
#use pprint to pretty-print a list
import pprint
pprint.pprint(lines[:5])
````

````python
#the first element is empty, so we remove it:
lines=lines[1:]
pprint.pprint(lines[:5])
````

````python
#convert this to list by separating at '\t':
dd=[q.split('\t') for q in lines]
pprint.pprint(dd[:3])
````

````python
# At this point you can do many of the data analysis 
# by working directly with the list we've just created, 
# e.g. using list comprehensions and numpy, etc. 
# This is useful for debugging, but could be cumbersome for non-trivial tasks,
# and slow for large datasets.
#
# Pandas is a python package for handling tabular data, which has methods 
# for data input-output and to various formats.
#
# First open the data file and find the number of lines in the header (remember that header ends with a '*/')

i=0 # counter is defined outside the scope of the "with" block
for line in open('../data/boknis.txt', 'r'): #read one line at a time
    if line.find('*/')>-1: break # if there's '*/' on a line, exit loop and close the file
    i+=1 # increment counter
print(i)
````
